#### _a 변수의 값을 가져오면서 a와는 다른 주소를 가리키도록 만드는 방법은?_

##### __1. [:]사용__

```python
a = [1,2,3]
```

```python
b = a[:]    # a의 리스트 값을 바꾸더라도 b리스트에는 영향을 끼치지 않는다
```

```python
a[1]=4
```

```python
a
```

```python
b
```

```python
b
```
