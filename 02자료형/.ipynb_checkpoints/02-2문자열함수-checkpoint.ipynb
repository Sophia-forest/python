## __문자열 관련 함수__

##### 사용방법 : 변수 이름 뒤에 '.'을 붙인 다음에 함수 이름을 써주면 된다

### __문자 개수 세기(count)__

```python
a = "hobby"
a.count("b")
```

##### 

### __위치 알려주기1(find)__

```python
a = "Python is the best choice"
```

```python
a.find('b')
```

```python
a.find("k")    # 찾는 문자나 문자열이 존재하지 않는다면 -1을 반환
```

##### 

### __위치 알려주기2(index)__

```python
a = "Life is too short"
```

```python
a.index('t')
```

```python
a.index('k')    # k가 없기 때문에 오류 발생
```

##### 

### __문자열 삽입(join)__

```python
",".join('abcd')
```

```python
",".join(['a','b','c','d'])
```

##### 

### __소문자를 대문자로 바꾸기(upper)__

```python
a = "hi"
a.upper()
```

##### 

### __대문자를 소문자로 바꾸기(lover)__

```python
a = "HI"
a.lower()
```

##### 

### __왼쪽 공백 지우기(lstrip)__

```python
a = "hi"
a.lstrip()
```

##### 

### __오른쪽 공백 지우기(rstrip)__

```python
a = "hi"
a.rstrip()
```

##### 

### __양쪽 공백 지우기(strip)__

```python
a="hi"
a.strip()
```

##### 

### __문자열 바꾸기(replace)__

```python
a = "Life is too short"
a.replace("Life", "Your leg")
```

###### replace(바뀌게 될 문자열, 바꿀 문자열)처럼 사용해서 문자열 안의 특정한 값을 다른 값으로 치환해 준다.

##### 

### __문자열 나누기(split)__

```python
a = "Life is too short"
```

```python
a.split()
```

```python
b = "a:b:c:d"
```

```python
b.split(':')
```
