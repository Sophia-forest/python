```python
grade['julliet']    # Key가 'julliet'인 딕셔너리의 Value를 반환
```

#### __딕셔너리 관련 함수__

##### __Key리스트 만들기(keys)__

```python
a = {'name':'pey', 'phone':'0119993323','birth':'1118'}
```

```python
a.keys()
```

######

```python
for k in a.keys():
    print(k)
```

######

```python
list(a.keys())
```

##### __Key로 Value 얻기(get)__

```python
a = {'name':'pey', 'phone':'0119993323','birth':'1118'}
```

```python
a.get('name')
```

```python
a.get('phone')
```

######

```python
a = {'name':'pey','phone':'0119993323','birth':'1118'}
```

```python
print(a.get('nokey'))    # None을 리턴함
```

```python
print(a['nokey'])
```

######

```python
a.get('foo','bar')
```

```python
# a 딕셔너리에는 'foo'에 해당하는 값이 없으므로 디폴트 값인 'bar'을 돌려준다.
```

######

##### __해당 Key가 딕셔너리 안에 있는지 조사하기(in)__

```python
a = {'name':'pey','phone':'0119993323', 'birth':'1118'}
```

```python
'name' in a
```

```python
'email' in a
```
