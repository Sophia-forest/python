### __8진수(Octal)와 16진수__

#### 8진수(Octal)를 만들기 위해서는 숫자가 0o또는 0O  로 시작하면 된다.

```python
a = 0o177
```

```python
a
```

#### 16진수(Hexadecimal)를 만들기 위해서는 0x로 시작하면 된다.

```python
a = 0x8ff
b = 0xABC
```

```python
a
```

```python
b
```
