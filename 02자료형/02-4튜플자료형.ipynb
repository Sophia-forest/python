### __튜플 자료형__

##### __리스트는 그 값의 생성,삭제,수정이 가능하지만 튜플은 그 값을 바꿀 수 없다.__

```python
t1 = ()
```

```python
t2 = (1,)
```

```python
t3 = (1,2,3)
```

```python
t4 = 1,2,3
```

```python
t5 = ('a', 'b', ('ab', 'cd'))
```

######

##### __1. 튜플 요솟값을 삭제하려 할 때__

```python
t1 = (1,2,'a','b')
```

```python
del t1[0]
```

```python
# 튜플은 요소를 지우는 행위가 지원되지 않는다.
```

######

##### __2. 튜플 요솟값을 변경하려 할 때__

```python
t1 = (1,2,'a','b')
```

```python
t1[0] = 'c'
```

```python
# 요솟값 변경도 오류 발생
```
