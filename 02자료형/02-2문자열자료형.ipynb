### __이스케이프 코드__ 
##### 출력물을 보기좋게

#### __\n__ : 문자열 안에서 줄을 바꿀 때 사용

#### __\t__ : 문자열 사이에 탭 간격을 줄 때 사용

#### __\\__ : 문자 \를 그대로 표현할 때 사용

#### __\'__ : 작은따옴표(')를 그대로 표현할 때 사용

#### __\r__ : 캐리지 리턴(줄 바꿈 문자, 현재 커서를 가장 앞으로 이동)

#### __\f__ : 폼 피드(줄 바꿈 문자, 현재 커서를 다음 줄로 이동)

#### __\a__ : 벨 소리(출력할 때 PC스피커에서 '삑'소리가 난다)

#### __\b__ : 백스페이스

#### __\000__ : 널 문자

#### __포맷 코드와 숫자 함께 사용하기__

#### __1. 정렬과 공백__

```python
"%10s" % "hi"
```

##### %10s : 전체 길이가 10개인 문자열 공간에서 대입되는 값을 오른쪾으로 정렬하고 그 앞의 나머지는 공백으로 남겨 두라는 의미

```python
"%-10sjane" % 'hi'
```

##### hi를 왼쪽으로 정렬하고 나머지는 공백으로 채움

#### __2. 소수점 표현하기__

```python
"%0.4f" % 3.42134234
```

```python
"%10.4f" % 3.42134234
```

#### __format함수를 사용한 포매팅__

##### 숫자 바로 대입하기 

```python
"I eat {0} apples" .format(3)
```

##### 문자열 바로 대입하기

```python
"I eat {0} apples".format("five")
```

##### 숫자 값을 가진 변수로 대입하기

```python
number = 3
```

```python
"I eat {0} apples".format(number)
```

##### 2개 이상의 값 넣기

```python
number = 10
```

```python
day = "three"
```

```python
"I ate {0} apples. so I was sick for {1} days.".format(number, day)
```

##### 이름으로 넣기

```python
"I ate {number} apples. so I was sick for {day} days.".format(number=10, day=3)
```

##### 인덱스와 이름을 혼용해서 넣기

```python
"I ate {0} apples. so I was sick for {day} days.".format(10,day=3)
```

##### 왼쪽 정렬

```python
"{0:<10}".format("hi")
```

##### 오른쪽 정렬

```python
"{0:>10}".format("hi")
```

##### 가운데 정렬 ( ^ : 가운데 정렬 / 빈 공간을 =로 채움 )

```python
"{0:=^10}".format("hi")
```

```python
"{0:!<10}".format("hi")         # 왼쪾(<)으로 정렬하고, 빈 공간을 ! 문자로 채움
```

##### 소수점 표현하기

```python
y = 3.421234234
```

```python
"{0:0.4f}".format(y)        # 소수점 4자리까지만 표현하기
```

```python
"{0:10.4f}".format(y)
```

##### { 또는 } 문자 표현하기

```python
"{{and}}".format()    # 
```

#### __f문자열 포매팅__

```python
name = '홍길동'
```

```python
age = 30
```

```python
f'나의 이름은 {name}입니다. 나이는 {age}입니다.'
```

```python
age = 30
```

```python
f'나는 내년이면 {age+1}살이 된다.'
```

```python
d = {'name':'홍길동', 'age':30}
```

```python
f'나의 이름은 {d["name"]}입니다. 나이는 {d["age"]}입니다.'
```

```python
f'{"hi":<10}'    # 왼쪽 정렬
```

```python
f'{"hi":>10}'    # 오른쪽 정렬
```

```python
f'{"hi":^10}'    # 가운데 정렬
```

```python
f'{"hi":=^10}'    # 가운데 정렬하고 =문자로 공백 채우기
```

```python
f'{"hi":!<10}'    # 왼쪽 정렬하고 ! 문자로 공백 채우기
```

```python
y = 3.42134234
```

```python
f'{y:0.4f}'    # 소수점 4자리까지만 표현
```

```python
f'{y:10.4f}'    # 소수점 4자리까지 표현하고 총 자릿수를 10으로 맞춤
```
