#### __집합 자료형의 특징__

##### _중복을 허용하지 않는다_ / _순서가 없다(Unordered)_

######

```python
s1 = set([1,2,3])
```

```python
l1 = list(s1)    # 리스트로 변환
```

```python
l1
```

```python
l1[0]
```

```python
t1 = tuple(s1)
```

```python
t1
```

```python
t1[0]
```

######

#### __교집합, 합집합, 차집합 구하기__

```python
s1 = set([1,2,3,4,5,6])
```

```python
s2 = set([4,5,6,7,8,9])
```

######

##### __교집합__

```python
s1 & s2
```

```python
s1.intersection(s2)
```

######

##### __합집합__

```python
s1|s2
```

```python
s1.union(s2)
```

######

##### __차집합__

```python
s1 - s2
```

```python
s2 - s1
```

```python
s1.difference(s2)
```

```python
s2.difference(s1)
```

######

#### __집합 자료형 관련 함수__

##### __값 1개 추가하기(add)__

```python
s1 = set([1,2,3])
```

```python
s1.add(4)
```

```python
s1
```

######

##### __값 여러 개 추가하기(update)__

```python
s1 = set([1,2,3])
```

```python
s1.update([4,5,6])
```

```python
s1
```

######

##### __특정 값 제거하기(remove)__

```python
s1 = set([1,2,3])
```

```python
s1.remove(2)
```

```python
s1
```
