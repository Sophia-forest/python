##### 주어진 자연수가 홀수인지 짝수인지 판별해 주는 함수(is_odd)

```python
def is_odd(number):
    if number%2==1:
        return True
    else:
        return False
```

##### 입력으로 들어오는 모든 수의 평균 값을 계산해 주는 함수를 작성해 보자(단, 입력으로 들어오는 수의 개수는 정해져 있지 않다)

```python
def avg_numbers(*args):
    result = 0
    for i in args:
        result += i
    return result / len(args)
```

```python
avg_numbers(1, 2)
```

```python
avg_numbers(1,2,3,4,5)
```
