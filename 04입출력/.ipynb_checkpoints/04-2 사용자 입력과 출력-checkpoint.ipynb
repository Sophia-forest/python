##### 프롬프트 값을 띄워서 사용자 입력받기

```python
# 사용자에게 입력받을 때, '숫자를 입력하세요'라든지 '이름을 입력하세요'라는 안내 문구 
# 또는 질문이 나오도록 하고 싶을 때가 있다. 그럴 때는 input()의 괄호 안에 
# 질문을 입력하여 프롬프트를 띄워주면 된다.
```

```python
input("질문 내용")
```

```python
number = input("숫자를 입력하세요:")
```

```python
print(number)
```

#### __print 자세히 알기__

```python
a = 123
print(a)
```

```python
a = "Python"
print(a)
```

#### __큰따옴표(")로 둘러싸인 문자열은 + 연산과 동일하다__

```python
print("Life" "is" "too short")
```

```python
print("Life"+"is"+"too short")
```

######

#### __문자열 띄어쓰기는 콤마로 한다__

```python
print("life", "is", "too short")
```
