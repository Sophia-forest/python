##### 곱하기/뺴기/나누기 기능 만들기

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first/self.second
        return result
```

```python
a = FourCal()
b = FourCal()
a.setdata(4,2)
b.setdata(3,8)
```

#### __생성자(Constructor)__

```python
a = FourCal()
a.add()
```

```python
# 파이썬 메소드 이름으로 __init__을 사용하면 이 메서드는 생성자가 된다.
class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    def setdata(self, first, second):
        self.first = first
        self.second = second
    def add(self):
        result = self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first/self.second
        return result
```

```python
def __init__(self, first, second):
    self.first = first
    self.second = second
```

```python
a = FourCal()    # first, second에 해당하는 값을 전달하여 객체를 생성해야 함 
```

```python
a = FourCal(4,2)
```

```python
print(a.first)
```

```python
print(a.second)
```

###### 

```python
a = FourCal(4,2)
```

```python
a.add()
```

```python
a.div()
```

##### 메서드 오버라이딩

```python
a = FourCal(4,0)
a.div()
```

```python
# 0으로 나눌 때 오류가 아닌 0을 돌려주도록 만들고 싶으면 ? 
```

```python
class SafeFourCal(FourCal):    # FourCal 클래스에 있는 div메서드를 동일한 이름으로 다시 작성
    def div(self):             # 부모 클래스(상속한 클래스)에 있는 메서드를 동일한 이름으로 다시 만드는 것
        if self.second == 0:   # 메서드 오버라이딩(Overriding , 덮어쓰기)
            return 0           # 부모 클래스의 메서드 대신 오버라이딩한 메서드가 호출
        else:
            return self.first / self.second
```

```python
a = SafeFourCal(4,0)    # 오류 처리함.
a.div()
```
