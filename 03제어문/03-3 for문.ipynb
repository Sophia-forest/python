#### __for문의 기본 구조__

##### __예제를 통해 for문 이해하기__

##### 1.전형적인 for문

```python
test_list = ['one','two','three']
```

```python
for i in test_list:
    print(i)
```

#####

##### 2.다양한 for문의 사용

```python
a = [(1,2), (3,4), (5,6)]
for (first, last) in a:
    print(first + last)
```

##### 3.for문의 응용

###### 총 5명의 학생이 시험을 보았는데 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 합격인지 불합격인지 결과를 보여 주시오

```python
marks = [90, 25, 67, 45, 80]
```

##### 

#### __for문과 continue문__

```python
marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks:
    number = number + 1
    if mark < 60: continue
    print("%d번 학생 축하합니다. 합격입니다." %number)
```

##### 

#### __for문과 함께 자주 사용하는 range함수__

```python
a = range(10)
a
```

```python
a = range(1,11)
a
```

###### __range함수의 예시 살펴보기__

```python
add = 0
for i in range(1, 11):
    add = add + i
print(add)
```

```python
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60: continue
    print("%d번 학생 축하합니다. 합격입니다." %(number+1))
```

###### 

###### __for와 range를 사용한 구구단__

```python
for i in range(2,10):
    for j in range(1,10):
        print(i*j, end=" ")
    print('')
```

###### 

#### __리스트 내포(리스트 안에 for문 포함하는) 사용하기__

```python
a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)
print(result)
```

###### 

```python
a = [1,2,3,4]
result = [num*3 for num in a]
print(result)
```

######

```python
a = [1,2,3,4]
result = [num*3 for num in a if num%2 == 0]
print(result)
```

###### 

```python
result = [x*y for x in range(2,10)
         for y in range(1,10)]
print(result)
```
