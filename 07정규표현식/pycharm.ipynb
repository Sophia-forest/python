```python
from urllib.parse import urlparse

myUrl = 'http://localhost:8090/api/v1/getrecord?city=seoul?zipcode=08206&mobile=01012341234&nickname=cat'

a = urlparse(myUrl)

print(a)

a = urlparse(myUrl).query

# print(a)

b = a.split('?')
print(b)

resultHtml = []

resultHtml.append(b[0])

print(resultHtml)

c = b[1].split('&')
print(c)

for i in range(len(c)):
    resultHtml.append(c[i])

print(resultHtml)

resultHtml2 = []
for i in range(len(resultHtml)):
    resultHtml2.append(resultHtml[i].replace("=",':'))    # 왜 이렇게 하면 안되는거지 ? # 공간을 새로 만들어야함.

resultHtml2

for i in range(len(resultHtml2)):
    print(resultHtml2[i])
```
