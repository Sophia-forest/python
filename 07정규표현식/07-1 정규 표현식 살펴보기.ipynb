```python
# 정규 표현식(Regular Expressions)은 복잡한 문자열을 처리할 때 사용하는 기법
# 파이썬만의 고유 문법이 아니라 문자열을 처리하는 모든 곳에서 사용한다. 
```

```python
# 주민등록번호 뒷자리를 * 문자로 변경해 보자.
```

```python
data = """
park 800905-1049118
kim 700905-1059119
"""
```

```python
result = []
for line in data.split("\n"):   # 엔터 기준으로 나눔
    word_result = []
    for word in line.split(" "):   # spacebar 기준으로 나눔 # " " 공백 넣어줘야함. park, kim 은 word에 들어가있음.
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit(): # isdigit() : 문자열이 숫자인지 아닌지 ture,false로 반환
            word = word[:6]+"-"+"*******"  # 추출한 word에서 또 추출하고 새로 word를 정의해줌
        word_result.append(word)  # word_result에 위에서 정의한 word를 넣어줌/ word_result (800905-********)
    result.append(" ".join(word_result))   # result에 다시 넣어줌. 한칸띄우고 넣어줌.
print("\n".join(result)) # 엔터 넣어서 나눠줌.
```

```python
import re

data = """
park 800905-1049118
kim 700905-1059119
"""

pat = re.compile("(\d{6})[-]\d{7}")
print(pat.sub("\g<1>-*******", data))
```
