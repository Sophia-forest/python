#### 3. 파이썬은 무료이지만 강력
##### 프로그램의 전반적인 뼈대는 파이썬으로 만들고 , 빠른 실행 속도가 필요한 부분은 C로 만들어서 파이썬 프로그램 안에 포함시킨다. (다른 언어로 만든 프로그램을 파이썬 프로그램에 포함 시킬 수 있다.)

#### 4. 파이썬은 간결
##### 줄을 잘 맞춰야 실행이 된다. 

```python
# simple.py
languages = ['python', 'perl', 'c', 'java']

for lang in languages:
    if lang in ['python', 'perl']:
        print("%6s need interpreter" % lang)
    elif lang in ['c', 'java']:
        print("%6s need compiler" % lang)
    else:
        print("should not reach here")
```
